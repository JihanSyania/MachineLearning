## Data Visualization

```python
klib.dist_plot(df_iris)
```

```python
sns.countplot(x=df_iris['species'])
#seaborn
```

```python
df_iris.plot(kind='scatter', x='sepal_length', y='sepal_width')
```

```python
df_iris.plot(kind='scatter', x='petal_length', y='petal_width')
```

```python
sns.scatterplot(data=df_iris, x=df_iris['petal_length'], y=df_iris['petal_width'], hue=df_iris['species'])
```

```python
sns.distplot(df_iris['sepal_length'], bins=20)
plt.show()
```

```python
sns.kdeplot(x=df_iris['sepal_length'], hue=df_iris['species'], shade=True)
```

```python
sns.boxplot(data=df_iris, x='species', y='petal_length') # petal_width, sepal_length, sepal_width
plt.xlabel('Species')
plt.ylabel('Length in cm')
plt.show()
```

```python
sns.pairplot(data=df_iris)
```

```python
sns.pairplot(data=df_iris, hue='species')
plt.show()
```

```python
sns.heatmap(data=df_iris.corr(), annot=True)
plt.show()
#annot untuk menampilkan detail angka setiap warna
```
